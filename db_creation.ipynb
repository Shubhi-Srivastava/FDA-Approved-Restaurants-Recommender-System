```python
import pandas as pd
import numpy as np
import random
from numpy import random as np_rand
import matplotlib.pyplot as plt
```

```python
table_list = ['businesses', 'food_type' 'inspections', 'violations', 'customer_rating', 'address_info']

# Attributes of tables
business_att = ['business_id', 'business_name', 'restaurant_type']
food_type_att = ['business_name', 'cuisine']
inspections_att = ['inspection_id', 'business_id', 'inspection_date', 'inspection_score', 'inspection_type']
violations_att = ['violation_id', 'business_id', 'violation_description', 'risk_category']
customer_rating_att = ['business_id', 'business_name', 'rating']
addresses_att = ['business_address', 'business_city', 'business_state', 'business_postal_code', 'business_id', 'business_latitude', 'business_longitude']
```

```python
rest_type = ['Fine Dining', 'Cafe', 'Fast Food', 'Bars & Pubs', 'Food Truck', 'Family', "Drive Thru"]
cuisine = ['Italian', 'Mexican', 'Korean', 'Indian', 'American', 'Japanese', 'Thai']
```

```python
def clean_str(str_):
    name_list = str_.split(" ")
    
    for i, name in enumerate(name_list):
        name = name.split("'")[0]
        name_list[i] = name
        
    return " ".join(name_list)
```

```python
def insert_n_rows(f, df, table, table_att, n):
    w_str = "\n\nINSERT INTO " + table + "("
    for att in table_att[:-1]:
        w_str += att + ", "
    
    w_str += table_att[-1] + ")\nVALUES\n"
    f.write(w_str)
    for i in range(n):
        tup = "("
        for c in table_att[:-1]:
            val = df.loc[df.index[i], c]
            if isinstance(val, str):
                val = clean_str(val)
                tup += "'" + val + "', "
            else:
                tup += str(val) + ", "
        
        final_col_val = df.loc[df.index[i], table_att[-1]]
        
        if isinstance(final_col_val, str):
            final_col_val = clean_str(final_col_val)
            tup += "'" + final_col_val + "')"
        else:
            tup += str(final_col_val) + ")"
        
        if (i < n-1):
            f.write(tup + ",\n")
    
    f.write(tup + ";\n")
    
```

```python
def create_sql_file2(name, df, rows):
    with open(name, "w", encoding='utf-8') as f:
        for table in table_list:
            f.writelines("delete from " + table + ";\n")
            
        insert_n_rows(f, df, 'food_type', food_type_att, rows)
        
```

```python
def create_sql_file(name, df, rows):
    with open(name, "w", encoding='utf-8') as f:
        for table in table_list:
            f.writelines("delete from " + table + ";\n")
            
        insert_n_rows(f, df, 'businesses', business_att, rows)
        insert_n_rows(f, df, 'inspections', inspections_att, rows)
        insert_n_rows(f, df, 'violations', violations_att, rows)
        insert_n_rows(f, df, 'customer_rating', customer_rating_att, rows)
        insert_n_rows(f, df, 'address_info', addresses_att, rows)
```
